```python
import sys
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
#from dash import jupyter_dash

sys.path.append('/Users/rebecca/Desktop/Coding_Projects/Exploring-Chemical-Space')
import chem_space as cs
import chem_space.plotting as plot
import chem_space.dashboard as db

```

```python
file = '~/Documents/NN_library_5-14-24.xlsx'
```

```python
# Import file data
#df = pd.read_csv(file)
df = pd.read_excel(file, sheet_name='all_desc')

smiles = df['SMILES']
ids = df['LibraryID']


#df.dropna(axis=0, inplace=True)
df.drop(columns=['SMILES', 'LibraryID', 'type', 'ligand_class'], inplace=True)

# Clean data
cut = 0.7
corr = df.corr(method=r2_score)
drop_cols = []
for col in corr:
    if corr[col].max(axis=0) >= cut and corr[col].max(axis=0) != 1.0:
        drop_cols.append(col)

df.drop(columns=drop_cols, inplace=True)
df_sc = pd.DataFrame(StandardScaler().fit_transform(df))
df_sc.columns = df.columns
display(df_sc)
```

```python
method = 'umap'
cluster = 'dbscan'

spc = cs.ChemicalSpace()
clus = cs.Cluster()

spc.make_space(df_sc, method=method)
spc.df = clus.get_clusters(spc.df, method=cluster)
print(spc.df)

fig = plot.plotly(spc, method)
fig.show()
```
